#### In the last few decades, it has become apparent to scientists of several different disciplines that the average global temperature has been rising and continues to do so. There are several lines of evidence for this increase in global temperature including the rise of atmospheric concentrations of greenhouse gases, the rise of global sea level, the melting of the cyrosphere and shifting global weather patterns. While Earth's past climate has shifted from greenhouse to icehouse conditions and vice-versa, the rapidly increasing mean global temperature is staggering and has never been observed in the geologic record. The most likely due to the industrialization of our species that begna in the 1750s and was aggravated by the discovery of fossil fuels in the late ninteenth century. Our current industrial and agricultural practices continute to release large amounts of greenhouse gases(carbon dioxide, methane, water vapor) into the atmosphere, already acclerating the effect of increasing mean global temeprature.This document wil serve as a general summary on the previously mentioned lines of evidence and how they confirm climate cahnge is anthropogenic.

#### <span style="color:red">(1)</span> Climate on Earth is controlled by the Sun's released energy. Incident suinlight arrives to the Earth as shortwave radiation, mostly as ultraviolet rays, penetrating the Earth's atomospehre, where it is either absorbed or reflected. Absorbed radiation at the surface of Earth is then released as longwave radiation, comprised of infrared wavelengths. <span style="color:red">(2)</span>Longwave radiation is absorbed by greenhouse gases, preventing the radiation from escaping and generating heat in the process. This effect maintains a comfortable temperature for life in the lower atomsphere. Carbon dioxide, methane, water vapor, nitrous oxides and fluorinated gases are considered greenhouse gases because the molecules are able to absorb infrared radiation. Of these 5 compounds, carbon dioxide has the greatest effect on climate forcing. Molecular nitrogen and oxygen in the atmosphere cannot absorb infrared radiation therefore they do not contribute to the greenhouse effect regardless of their concentration

  


#### **Figure 1.** Simplified diagram of the greenhouse effect. Incident solar energy, shortwave radiation, penetrates through Earth's  atmosphere where it is either absorbed by the surface or reflected. The shortwave radiation absorbed by the Earth is released as longwave radiatiion. Longwave radiation will radiate back into space, but some is absorbed by greenhouse gasses in the lower atmosphere. The greenhouse gases increase the heat retained near the Earth's surface. Figure obtained from [Wikimedia Commons.](https://commons.wikimedia.org/wiki/Category:Greenhouse_effect#/media/File:Greenhouse_effect_model.svg)

#### <span style="color:red">(5)</span> Before the onset of anthropogenic climate change, Earth's atmospheric CO2 varied approximately between 200 during glacial periods, and 280 ppm during interglacial periods. Drilled ice cores from Antartica and Greenland provide a record of atmospheric CO2 preserved as trapped bubbles wihtin the ice. The bubbles can be extracted and concentrations be measured to extrapolate the global climate. The oldest record was taken from the [Vostok ice core](https://climate.nasa.gov/climate_resources/24/graphic-the-relentless-rise-of-carbon-dioxide/#:~:text=They%20tell%20us%20that%20levels,see%20fluctuations%20in%20the%20graph) in Antartica.

#### <span style="color:red">(7)</span> Each greenhouse gas is released into the atmosphere by different sources. Methane is predominantly generated by the transport of coal, natural gas & oil while carbon dioxide is generated by burning fossil fuels. Nitrous oxide is mostly produced by agriculture by the way of fertilized soil and animal waste. Methane is 80x more powerful than carbon dioxide when generating heat, though it last much less in the atmosphere than CO2. Carbon dioxide can remain in the atmosphere for up to 1000 years. Additional information can obtained [here.](https://www.epa.gov/ghgemissions/overview-greenhouse-gases#:~:text=Nitrous%20oxide%20(N2O,as%20during%20treatment%20of%20wastewater.)

Based on figure 4, nitrous oxide has been steadily increasing. Methane has increased steadily but more slowly than the latter two. Between the mid 1990s to mid 2000s, methane concentrations stabalized before shaply increasing again in 2007. It is not well understood why there was a brief pause admist the increasing trend of methane. Research suggests that the source is not burning of fossil fuels but the microbial generation of methane in wetlands due to the increasse in temperature and humidity. Areas of increased methane are seen in the tropics and mid latitude northern hemisphere.
